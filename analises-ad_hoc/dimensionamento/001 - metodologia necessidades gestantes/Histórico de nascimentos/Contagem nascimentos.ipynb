# Contagem do número de nascimentos por data e município de residência

Contagem da base SINASC 

```python
sql_total = '''SELECT * FROM(SELECT SUBSTRING(CODMUNRES, 1, 2) AS UF, CODMUNRES, DTNASC, 
                             COUNT(*) AS contagem FROM Dados.sinasc.DN
                             GROUP BY CODMUNRES, DTNASC)
                             WHERE UF = '52' '''

dataframe1 = pd.read_sql(sql_total,cnxn)
dataframe1.head()
```

Contagem da base SINASC 2020 - Preliminar

```python
sql_2020 = '''SELECT * FROM(SELECT SUBSTRING(CODMUNRES, 1, 2) AS UF, CODMUNRES, DTNASC, 
                                 COUNT(*) AS contagem FROM Dados.sinasc.preliminar.sinasc_2020
                                 GROUP BY CODMUNRES, DTNASC)
                                 WHERE UF = '52' '''

dataframe2020 = pd.read_sql(sql_2020,cnxn)
dataframe2020.head()
```

Contagem da base SINASC 2021 - Preliminar

```python
sql_2021 = ''' SELECT * FROM(SELECT SUBSTRING(CODMUNRES, 1, 2) AS UF, CODMUNRES, DTNASC, 
                                       COUNT(*) AS contagem FROM Dados.sinasc.preliminar.sinasc_2021                                                    GROUP BY CODMUNRES, DTNASC)
                                       WHERE UF = '52' '''

dataframe2021 = pd.read_sql(sql_2021,cnxn)
dataframe2021.head()
```

```python
df = pd.concat([dataframe1,dataframe2020,dataframe2021])
df.head()
```
